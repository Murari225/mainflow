```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'householdtask3.csv'
data = pd.read_csv(file_path)

# Scatter Plot: Age vs. Income
plt.figure(figsize=(10, 6))
plt.scatter(data['age'], data['income'], alpha=0.5, edgecolors='w', s=100)
plt.title('Scatter Plot: Age vs. Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.grid(True)
plt.show()

# Bar Plot: Year vs. Total Households
plt.figure(figsize=(10, 6))
plt.bar(data['year'], data['tot_hhs'], color='skyblue')
plt.title('Bar Plot: Year vs. Total Households')
plt.xlabel('Year')
plt.ylabel('Total Households')
plt.xticks(data['year'].unique())
plt.grid(axis='y')
plt.show()

# Histogram: Household Income
plt.figure(figsize=(10, 6))
plt.hist(data['income'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Histogram: Household Income')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
